## ðŸŸ¦ 2ï¸âƒ£ Import Required Libraries

```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
from wordcloud import WordCloud
import re
import warnings
warnings.filterwarnings("ignore")
```

## ðŸŸ¦ 3ï¸âƒ£ Load the Dataset

```

df = pd.read_csv("chatgpt_reviews.csv")
df.head()
```


## ðŸŸ¦ 4ï¸âƒ£ Data Overview

The dataset contains:
- Review ID
- Review Text
- Rating (0â€“5)
- Review Date


```

df.info()
df.isnull().sum()
```

## ðŸŸ¦ 5ï¸âƒ£ Data Cleaning & Preprocessing

```

df = df.dropna(subset=['Review'])
df['Review date'] = pd.to_datetime(df['Review date'])
df['Month'] = df['Review date'].dt.to_period('M')
df['Review Length'] = df['Review'].apply(len)
df.head()
```

## ðŸŸ¦ 6ï¸âƒ£ Rating Distribution Analysis

```

plt.figure()
sns.countplot(x='Rating', data=df)
plt.title("Rating Distribution")
plt.show()

print("Average Rating:", round(df['Rating'].mean(),2))
```

## ðŸŸ¦ 7ï¸âƒ£ Sentiment Analysis

```

def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

df['Sentiment Score'] = df['Review'].apply(get_sentiment)

def sentiment_label(score):
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

df['Sentiment'] = df['Sentiment Score'].apply(sentiment_label)

sns.countplot(x='Sentiment', data=df)
plt.title("Sentiment Distribution")
plt.show()
```

## ðŸŸ¦ 8ï¸âƒ£ Word Cloud Analysis

```

positive_text = " ".join(df[df['Sentiment']=="Positive"]['Review'])
wordcloud = WordCloud(width=800, height=400).generate(positive_text)

plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Positive Reviews Word Cloud")
plt.show()
```

## ðŸŸ¦ 9ï¸âƒ£ Time-Series Analysis

```

monthly_reviews = df.groupby('Month').size()

monthly_reviews.plot()
plt.title("Reviews Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Reviews")
plt.show()

monthly_rating = df.groupby('Month')['Rating'].mean()

monthly_rating.plot()
plt.title("Average Rating Over Time")
plt.xlabel("Month")
plt.ylabel("Average Rating")
plt.show()
```

## ðŸŸ¦ ðŸ”Ÿ Issue Identification (Negative Reviews)

```

from collections import Counter

negative_words = " ".join(df[df['Sentiment']=="Negative"]['Review']).lower()
words = re.findall(r'\b\w+\b', negative_words)

common_words = Counter(words).most_common(10)
common_words
```
